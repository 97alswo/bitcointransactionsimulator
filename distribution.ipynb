```python
import blocksci
import pandas as pd
from datetime import datetime, timedelta
import pickle

chain = blocksci.Blockchain("conf.json")

dx = pd.read_excel('data/data.xlsx')
datar = []
for i in range(len(dx)):
    tmp = [dx["ID"][i], dx["Value"][i], dx["Date"][i]]
    ttmp = datetime.strptime(str(tmp[2]), "%Y%m%d")
    tmp[2] = ttmp
    datar.append(tmp)
data = []
for v in datar:
    if v not in data:
        data.append(v)
data_t = []
for d in data:
    data_t.append(d[2])
start = min(data_t)
end = max(data_t)

block = chain.range(start, end)

cm = blocksci.cluster.ClusterManager("cluster4", chain)
```

```python
count = 0
for co in ctxo:
    coa = co.address
    if coa.full_type != "nulldata":
        if coa.first_tx.index == co.tx_index:
            count += 1
print(count)
```

```python
c_address = []
for co in ctxo:
    coa = co.address
    if coa.full_type != "nulldata":
        if coa not in c_address:
            c_address.append(coa)
print(len(c_address))
```

```python
#miner cluster
cil = []
for co in ctxo:
    coa = co.address
    if coa.full_type != "nulldata":
        cluster = cm.cluster_with_address(coa)
        ci = cluster.index
        if ci not in cil:
            cil.append(ci)
print(len(cil))
```

```python
#mine reward distribution
co_value = []
for co in ctxo:
    if co.value != 0:
        co_value.append(co.value)
print(len(co_value))
```

```python
mining = 0
relay = 0
peel = 0
distribution = 0
sweep = 0
mimo = 0

for tx in txs:
    count += 1
    txic = tx.input_count
    txoc = tx.output_count
    
    if txic == 0:
        mining += 1
        continue
    else : 
        if txoc == 2:
            peel += 1
            continue
        else : 
            if txic == 1:
                if txoc == 1:
                    relay += 1
                    continue
                else :
                    distribution += 1
                    #distribution_output.append(txoc)
                    continue
            else : 
                if txoc == 1:
                    sweep += 1
                    #sweep_input.append(txic)
                    continue
                else : 
                    mimo += 1
                    #mimo_io.append((txic, txoc))

print("mining : " + str(mining))
print("relay : "+str(relay))
print("peel : "+str(peel))
print("distribution : "+str(distribution))
print("sweep : "+str(sweep))
print("mimo : "+str(mimo))
print((mining + relay + peel + distribution + sweep + mimo == len(txs.input_count)))
```
